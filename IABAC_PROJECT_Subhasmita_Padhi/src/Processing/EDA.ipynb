## Encoding of data

```python
# load data
# creating a list of only numerical values
num = [u'Age', u'DistanceFromHome', u'EmpEducationLevel', u'EmpEnvironmentSatisfaction', u'EmpHourlyRate'
 , u'EmpJobInvolvement', u'EmpJobLevel', u'EmpJobSatisfaction'
 , u'NumCompaniesWorked', u'EmpLastSalaryHikePercent'
 , u'EmpRelationshipSatisfaction', u'TotalWorkExperienceInYears'
 , u'TrainingTimesLastYear', u'EmpWorkLifeBalance'
 , u'ExperienceYearsAtThisCompany', u'ExperienceYearsInCurrentRole'
 , u'YearsSinceLastPromotion', u'YearsWithCurrManager']

# Empty list to store columns with categorical data
categorical = []
for col, value in df.iteritems():
    if value.dtype == 'object':
        categorical.append(col)

# Store the numerical columns in a list numerical
numerical=df.columns.difference(categorical)
# Store the categorical data in a dataframe called df_cat
df_cat = df[categorical]
df_cat = df_cat.drop(['EmpNumber'], axis=1)

# Store the numerical features to a dataframe df_num
df_num = df[numerical]

df_cat = pd.get_dummies(df_cat)
# Concat the two dataframes together columnwise
df_final = pd.concat([df_num, df_cat], axis=1)
df_final.shape

df_final1=df_final.drop(['PerformanceRating'], axis=1)


X = df_final1.iloc[:,:]

Y =df_final.loc[:,['PerformanceRating']]
df_final.head()
```

- All catagorical data encoded in to Integer value.

## Top 3 Important Factors effecting employee performance
  

To find out top 3 feature that effecting employee performance few methods are there.Here ExtraTreesClassifier use to determine.

```python

# feature extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
print(model.feature_importances_)
x_features=df_final1.columns[:]
print(sorted(zip(model.feature_importances_, x_features), reverse=True)[:3])

```

## 1.Gender vs PerformanceRating

```python
dept_crosstab=pd.crosstab(df['Gender'],df['PerformanceRating'], margins=True)
print(dept_crosstab)
pd.crosstab(df['Gender'],df['PerformanceRating']).plot.bar(stacked=True,color=('blue','brown','green'))
plt.legend(title='Performance')
plt.show()
```

### Chisquare test between gender and performanceRating if they are depended or not
- H0 = Gender and performanceRating is independent
- H1 = Gender and performanceRating is dependent


```python
Gender_chi_2,Gender_p_val,Gender_deg_free,Gender_freq_count=stats.chi2_contingency(pd.crosstab(df['Gender'],df['PerformanceRating']))
print("Chi square value : ",Gender_chi_2)
print("P value : ",Gender_p_val)
print("Degree of freedom : ",Gender_deg_free)
print("Expected Frequency Count matrix : ")
print(Gender_freq_count)
```

Observation:
    - If p value < 0.05 then null Hypothesis or H0 is False.
    - Here P value > 0.05 so H0 true
    - Gender and performanceRating is independent

# C:  Continuous Vs Catagorical Data

### 1. Boxplot and Annova test Between DistanceFromHome and PerformanceRating

```python
plt.figure(figsize=(6,4))
sns.boxplot(y='DistanceFromHome',x='PerformanceRating',data=df)
```

### Annova Test 
Our assumption for all Continuous same
- H0 = Mean between the diffrent groups are not very Large
- H1 = Mean between the diffrent groups are very large

```python
pr=df.groupby('PerformanceRating')
#Performance Rating 2 group
pr_2=pr.get_group(2)
#Performance Rating 3 group
pr_3=pr.get_group(3)
#Performance Rating 4 group
pr_4=pr.get_group(4)
stats.f_oneway(pr_2['DistanceFromHome'],pr_3['DistanceFromHome'],pr_4['DistanceFromHome']) # Annova Testing
```

### Observation
BOX PLOT

 1. pr_1 = Performance rating of catagory 1  
 2. pr_2 = Performance rating of catagory 2
 3. pr_3 = Performance rating of catagory 3
 4. pr_4 = Performance rating of catagory 4 
 5. IQR = InterQuartileRange
 6. Q1 = Lower Quartile(25%)
 7. Q3 = Upper Quartile(75%)
    
- The above box plot Shows that there is some outlier in performanceRating 4.
- Pr_2(IQR)> Pr_3(IQR)> pr_4(IQR)
- pr_2(Q1) = pr_3(Q1) = pr_4(Q1)
- pr_2(Q3) > pr_3(Q3) > pr_4(Q3)

ANNOVA TEST
- if P value less then 0.05 then H0 rejected and H1 will be considerabe. 
- Here p value is greater then 0.05 so H0 is true.
- Mean between the diffrent groups are not very large
